```python
# Cargar el archivo geojson
barrios_geojson = gpd.read_file('data/neighbourhoods.geojson')

# Anadimos una columna price_int en df_listing2 para poder hacer la union con df_preciomedio
df_listing2['price_int'] = df_listing2['price'].str.replace('$', '').str.replace(',', '').astype(float)

# Eliminamos todas lineas que tengan en la columna price_int un valor mayo a 2500
df_listing2 = df_listing2[df_listing2['price_int'] <= 2500]

# Convertimos los valores del df_preciomedio a float para el a√±o 2024
df_preciomedio['2024'] = df_preciomedio['2024'].replace('-', '0').astype(float)

# Agrupar la informaci√≥n por barrio
barrios_grouped = df_listing2.groupby('neighbourhood_cleansed').agg({
    'latitude': 'mean',
    'longitude': 'mean',
    'id': 'count',
    'price_int': 'mean'
}).reset_index()

# Unir los datos de precio medio con barrios_grouped
barrios_grouped = barrios_grouped.merge(df_preciomedio[['Territorio', '2024']], left_on='neighbourhood_cleansed', right_on='Territorio', how='left')

# Crear un mapa centrado en Barcelona
m = folium.Map(location=[41.3851, 2.1734], zoom_start=13)

# Agregar los barrios al mapa
folium.GeoJson(
    barrios_geojson,
    name='neighbourhoods',
    style_function=lambda feature: {
        'fillColor': '#blue',
        'color': 'black',
        'weight': 1,
        'dashArray': '5, 5'
    }
).add_to(m)

# Agregar marcadores para cada barrio con el precio medio
for _, row in barrios_grouped.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"Barrio: {row['neighbourhood_cleansed']}, Pisos: {row['id']}, Precio Medio: {round(float(row['2024']), 2)}, Precio Medio Turistico: {round(row['price_int'], 2)}",
        tooltip=f"Barrio: {row['neighbourhood_cleansed']}, Pisos: {row['id']}, Precio Medio: {round(float(row['2024']), 2)}, Precio Medio Turistico: {round(row['price_int'], 2)}",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(m)

# Mostramos el mapa interactivo
m
```

```python
import matplotlib.pyplot as plt

# Filtrar los datos para incluir solo el tipo de territorio "Municipi"
df_municipi = df_preciomedio[df_preciomedio['Tipo de territorio'] == 'Municipi']

# Seleccionar las columnas de a√±os y convertirlas a n√∫meros
df_municipi_evol = df_municipi.melt(id_vars=['Territorio'], var_name='A√±o', value_name='Precio', ignore_index=False)
df_municipi_evol = df_municipi_evol[df_municipi_evol['A√±o'].str.isdigit()]  # Filtrar solo columnas de a√±os
df_municipi_evol['A√±o'] = pd.to_numeric(df_municipi_evol['A√±o'], errors='coerce')
df_municipi_evol['Precio'] = pd.to_numeric(df_municipi_evol['Precio'], errors='coerce')

# Eliminar valores nulos
df_municipi_evol = df_municipi_evol.dropna()

# Filtrar los √∫ltimos 10 a√±os
ultimo_ano = df_municipi_evol['A√±o'].max()
df_municipi_evol = df_municipi_evol[df_municipi_evol['A√±o'] >= (ultimo_ano - 10)]

# Crear el gr√°fico
plt.figure(figsize=(10, 6))
for municipio in df_municipi_evol['Territorio'].unique():
    data = df_municipi_evol[df_municipi_evol['Territorio'] == municipio]
    plt.plot(data['A√±o'], data['Precio'], label=municipio)

# Configurar el gr√°fico
plt.title('Evoluci√≥n del Precio por A√±o (√öltimos 10 A√±os, Tipo de Territorio: Municipi)', fontsize=14)
plt.xlabel('A√±o', fontsize=12)
plt.ylabel('Precio Medio', fontsize=12)
plt.legend(title='Territorio', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```

```python
# quiero comparar en un grafico la columna license de los dataframe df_listing2, df_listing_240615, df_listing_240320 y df_listing_231213

import seaborn as sns

# Creamos un diccionario para iterar f√°cilmente por los dataframes
dataframes = {
    'Sep-24': df_listing2,
    'Jun-24': df_listing_240615,
    'Mar-24': df_listing_240320,
    'Dec-23': df_listing_231213
}
# En esta lista almacenaremos las estad√≠sticas de cada dataframe
registros = []

for nombre, df in dataframes.items():
    # Conteo de valores "Exempte"
    exempte_count = df['license'].eq('Exempt').sum()

    # Conteo de valores faltantes
    missing_count = df['license'].isna().sum()

    # Valores diferentes (excluyendo NA y "Exempte")
    valores_unicos = df['license'].dropna().unique().tolist()
    if 'Exempte' in valores_unicos:
        valores_unicos.remove('Exempt')
    different_count = len(valores_unicos)

    registros.append([nombre, different_count, missing_count, exempte_count])

# Creamos un DataFrame con las estad√≠sticas
df_plot = pd.DataFrame(registros, columns=['DataFrame', 'Valores Diferentes', 'Valores Faltantes', 'Valor Exempt'])

# Reestructuramos el DataFrame para facilitar el plot con seaborn
df_plot_melted = df_plot.melt(id_vars='DataFrame',
                              value_vars=['Valores Diferentes', 'Valores Faltantes', 'Valor Exempt'],
                              var_name='Categoria', value_name='Conteo')

# Graficamos
plt.figure(figsize=(15,6))
sns.barplot(x='DataFrame', y='Conteo', hue='Categoria', data=df_plot_melted)
plt.title('Comparaci√≥n de valores en la columna "license"')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show() 
```
